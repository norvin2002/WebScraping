```python
# import libraries
import requests
import pandas as pd
import time
import random
import bs4
```

```python
test_url = "https://www.channelnewsasia.com/news/topic/crime"
```

```python
headers = {"User-Agent" : "Mozilla 5.0/"} #only if you get error 403 while running requests without header set

test_res = requests.get(test_url, headers = headers)
#test_res = requests.get(test_url)
```

```python
# raw text
test_res.text
```

```python
# create a BeautifulSoup object
test_soup = bs4.BeautifulSoup(test_res.text)
test_soup
```

```python
# ol --> ordered list in HTML
# retrieve ol tag
# .find gets the first tag that fits the description
# class is a reserved keyword thus if it needs to be used in a function, underscore is needed
test_ol = test_soup.find('ol', class_ = 'result-section__list')
```

```python
test_ol
```

```python
# go through test_ol and get all li tags and put them into a list
# li --> list item
test_li_list = test_ol.find_all('li', class_ = 'result-section__list-item')

test_li_list
```

```python
len(test_li_list)
```

```python
# take first item in list
test_first = test_li_list[0]
test_first

# href = the url to go into specific page
```

```python
# try to retrieve category
test_category = test_first.find("a", class_ = "teaser__category").text
test_category
```

```python
# retrieve title
test_title = test_first.find("a", class_ = "teaser__title").text
test_title
```

```python
# retrieve URL
# use square bracket to retrieve info in tag

test_first.find("a", class_ = "teaser__title")['href']
```

```python
# retrieve time

test_first.find('time')['datetime']
```

```python
categories = []
titles = []
urls = []
timestamps = []

# loop through all ten items
for item in test_li_list:
    temp_category = item.find("a", class_ = "teaser__category").text
    categories.append(temp_category)
    
    temp_title = item.find('a', class_ = 'teaser__title').text
    titles.append(temp_title)
    
    temp_url = 'https://www.channelnewsasia.com' + item.find('a', class_ = 'teaser__title')['href']
    urls.append(temp_url)
    
    temp_timestamp = item.find('time')['datetime']
    timestamps.append(temp_timestamp)
    
    
```

```python
temp_df = pd.DataFrame({"category": categories,
                       "title":titles,
                       "url": urls,
                       "timestamp": timestamps})
temp_df
```

```python
test_first_url = temp_df['url'][7]
```

```python
test_article_soup = bs4.BeautifulSoup(test_article_res.text)
test_article_soup
```

```python
test_article_p = test_article_soup.find('div', class_ = 'c-rte--article')
```

```python
test_article_p
```

```python
clean_article_p = []


# strip white spaces. \xa0 is unicode space
for p in test_article_p:
    p.encode('unicode_escape')
    clean_article_p.append(p.text.strip().replace("\xa0", "").replace("\\", ""))
    
test_article = " ".join(clean_article_p)
test_article
```

```python
articles = []

for url in temp_df['url']:
    try:
        temp_article_res = requests.get(url, headers = headers)
        temp_article_soup = bs4.BeautifulSoup(temp_article_res.text)
        temp_article_p = temp_article_soup.find("div", class_ = "c-rte--article").find_all('p')

        temp_clean_article_p = []

        for p in temp_article_p:
            temp_clean_article_p.append(p.text.strip().replace("\xa0", " ").replace("\\",""))

        temp_article = " ".join(temp_clean_article_p)

        articles.append(temp_article)
        
    except AttributeError:
        articles.append("")
        
    time.sleep(random.randint(1,10))
```

```python
temp_df['articles'] = articles
temp_df
```

```python
# Loop through all pages

# get the ending page number
page_end = int(test_soup.find('p', class_ = 'pagination__index').find_all('strong', class_ = 'pagination__index-value')[1].text)
```

```python
page_end
```

```python
categories = []
titles = []
urls = []
timestamps = []

for page in range(page_end):
    
    base_url = "https://www.channelnewsasia.com/news/topic/crime?pageNum="
    test_url = base_url + str(page)
    
    # request
    test_res = requests.get(test_url, headers = headers)
    
    # create soup
    test_soup = bs4.BeautifulSoup(test_res.text)
    
    # get all list items
    test_li_list = test_soup.find('ol', class_ = 'result-section__list').find_all('li', class_ = 'result-section__list-item')
    

    # loop through all ten items
    for item in test_li_list:
        temp_category = item.find("a", class_ = "teaser__category").text
        categories.append(temp_category)

        temp_title = item.find('a', class_ = 'teaser__title').text
        titles.append(temp_title)

        temp_url = 'https://www.channelnewsasia.com' + item.find('a', class_ = 'teaser__title')['href']
        urls.append(temp_url)

        temp_timestamp = item.find('time')['datetime']
        timestamps.append(temp_timestamp)
        
    time.sleep(random.randint(1,10))
    
# create a dataframe with all news articles
temp_df = pd.DataFrame({"category": categories,
                   "title":titles,
                   "url": urls,
                   "timestamp": timestamps})
```

```python
temp_df
```

```python
# get content of each article
articles = []

for url in temp_df['url']:
    try:
        temp_article_res = requests.get(url, headers = headers)
        temp_article_soup = bs4.BeautifulSoup(temp_article_res.text)
        temp_article_p = temp_article_soup.find("div", class_ = "c-rte--article").find_all('p')

        temp_clean_article_p = []

        for p in temp_article_p:
            temp_clean_article_p.append(p.text.strip().replace("\xa0", " ").replace("\\",""))

        temp_article = " ".join(temp_clean_article_p)

        articles.append(temp_article)
    
    except AttributeError:
        articles.append("")
    
    time.sleep(random.randint(1,10))
```

```python
# add the content of each news
temp_df['articles'] = articles
temp_df
```
